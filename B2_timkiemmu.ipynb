# **1> Cài đặt thuật toán BFS**

Base on the code prof provides: Slide 19, 20.

```python
# As a graph
graph = {
    'A' : ['B', 'C'],
    'B' : ['D', 'E'],
    'C' : ['F', 'G'],
    'D' : ['H', 'I'],
    'E' : ['J', 'K'],
    'F' : ['L', 'M'],
    'G' : ['N', 'O'],
    'H' : [],
    'I' : [],
    'J' : [],
    'K' : [],
    'L' : [],
    'M' : [],
    'N' : [],
    'O' : []
}
```

```python
# For demo1
def BFS(graph, init, goal):
  frontier = [init]
  explored = []
  while frontier:
    state = frontier.pop(0)
    explored.append(state)
    if goal == state:
      return explored
    for neighbor in graph[state]:
      # print(neighbor)
      if neighbor not in (explored and frontier):
        frontier.append(neighbor)
  return False
```

```python
# For demo2
def BFS_tree(init: Node, goal):
  frontier = [init]
  explored = []
  while frontier:
    state = frontier.pop(0)
    explored.append(state)
    if goal == state.tag:
      return explored
    for neighbor in tree.children(state.identifier):
      # print(neighbor)
      if neighbor not in (explored and frontier):
        frontier.append(neighbor)
  return False
```

Base on the code prof provides: Slide 31, 32.

```python
#For demo3
def DFS(graph, init, goal):
  frontier = [init]
  explored = []
  while frontier:
    state = frontier.pop(len(frontier)-1)
    explored.append(state)
    if goal == state:
      return explored
    for neighbor in graph[state]:
      # print(neighbor)
      if neighbor not in (explored and frontier):
        frontier.append(neighbor)
  return False
```

# **3> Cài đặt thuật toán BFS và DFS trên cấu trúc dữ liệu tự xây dựng**

I choose adjacency list as a data structure to implement BFS and DFS algorithm because it can represent data well similar to graph or tree. In the otherhand, thanks to the adjacency list, we can develop a more complex graph to examine the performance of DFS or BFS algorithms.

I use defaultdict instead of regular dict due to its ability to auto-initialize keys.

```python
class Graph_from_adj_list:
  def __init__(self, nodes: int):
    # Store a adjacency list as a dictionary
    # The default dictionary would create an empty list as a default (value)
    # for the nonexistent keys.
    self.adjlist = defaultdict(list)
    self.nodes = nodes
  def AddEdge(self, src: str, dst: str):
    # src stands for source, dst stands for destination
    self.adjlist[src].append(dst)
    self.adjlist[dst].append(src)
  def Display(self):
    for item in self.adjlist.items():
      print(item)
  def BFS(self, init, goal):
    frontier = [init]
    explored = []
    while frontier:
      state = frontier.pop(0)
      explored.append(state)
      if goal == state:
        return explored
      for neighbor in self.adjlist[state]:
        # print(neighbor)
        if neighbor not in explored and neighbor not in frontier: # Difference
          frontier.append(neighbor)
    return False
  def DFS(self, init, goal):
    frontier = [init]
    explored = []
    while frontier:
        state = frontier.pop()
        if state not in explored:
            explored.append(state)
            if goal == state:
                return explored
            for neighbor in self.adjlist[state]:
                if neighbor not in explored and neighbor not in frontier:
                    frontier.append(neighbor)
    return False

  def DFS_Limited(self, init, goal, limit):
    frontier = [(init, 0)] # choose frontier for depth storage...
    explored = []
    while frontier:
        state, depth = frontier.pop()
        if depth <= limit:
            if state not in explored:
                explored.append(state)
                if goal == state:
                    return explored
                for neighbor in self.adjlist[state]:
                    if neighbor not in explored and neighbor not in [f[0] for f in frontier]:
                        frontier.append((neighbor, depth + 1))
    return False
```

```python
nodes = 15
g = Graph_from_adj_list(nodes)
g.AddEdge("A", "B")
g.AddEdge("A", "C")
g.AddEdge("B", "D")
g.AddEdge("B", "E")
g.AddEdge("C", "F")
g.AddEdge("C", "G")
g.AddEdge("D", "H")
g.AddEdge("D", "I")
g.AddEdge("E", "J")
g.AddEdge("E", "K")
g.AddEdge("F", "L")
g.AddEdge("F", "M")
g.AddEdge("G", "N")
g.AddEdge("G", "O")
print("Adjacency list:")
g.Display()
```

# **4> Cài đặt DFS + limited**

```python
# From the function for demo3, and limit that means the depth we need to search to.
def DFS_Limited(graph, init, goal, limit):
    frontier = [(init, 0)]
    explored = []
    while frontier:
        state, depth = frontier.pop()
        if depth <= limit:
            explored.append(state)
            if goal == state:
                return explored
            for neighbor in graph[state]:
                if neighbor not in explored and neighbor not in [f[0] for f in frontier]:
                    frontier.append((neighbor, depth + 1))
        else:
            continue
    return False
```
